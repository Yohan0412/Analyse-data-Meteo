# CorrÃ©lation entre TempÃ©rature et Consomation Gaz :

La corrÃ©lation est de -0. 53 ce qui signifie que plus le tempÃ©rature baisse plus la consomation de Gaz augmente 
, probablement en raison de lâ€™utilisation du chauffage.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[['temperature_2m']].values  # Variable explicative (2D)
y = df['conso_gaz_mw'].values        # Variable Ã  expliquer

model = LinearRegression()
model.fit(X, y)

print("Coefficient (impact tempÃ©rature) :", model.coef_[0])
print("Intercept :", model.intercept_)
print("Score RÂ² :", model.score(X, y))
```

## RÃ©gression linÃ©aire simple :

CoÃ©fficient : -261 ce qui signifie que pour chaque augmentation de 1Â°C la consomation baisse en moyenne de 261 Mw


Intercept : La valeur estimÃ©e de la consomation quand la tempÃ©rature est de 0Â°C est de 7606 Kw

Le Score R^2 indique de 27% de la variance de la consomation est expliquÃ©e par la tempÃ©rature

```python
import matplotlib.pyplot as plt
import pandas as pd

# S'assurer que datetime est bien au format datetime
df["datetime"] = pd.to_datetime(df["datetime"])
df["month"] = df["datetime"].dt.month


conso_gaz_moyenne = df.groupby("month")["conso_gaz_mw"].mean()
temp_moyenne = df.groupby("month")["temperature_2m"].mean()


fig, ax1 = plt.subplots(figsize=(10, 5))


ax1.bar(conso_gaz_moyenne.index, conso_gaz_moyenne, color="lightcoral", label="Conso Gaz (MW)")
ax1.set_xlabel("Mois")
ax1.set_ylabel("Consommation de gaz (MW)", color="lightcoral")
ax1.tick_params(axis='y', labelcolor="lightcoral")
ax1.set_xticks(range(1, 13))
ax1.set_xticklabels(["Jan", "FÃ©v", "Mar", "Avr", "Mai", "Juin", "Juil", "AoÃ»t", "Sept", "Oct", "Nov", "DÃ©c"])

# Courbe pour la tempÃ©rature 
ax2 = ax1.twinx()
ax2.plot(temp_moyenne.index, temp_moyenne, color="steelblue", marker="o", label="TempÃ©rature (Â°C)")
ax2.set_ylabel("TempÃ©rature moyenne (Â°C)", color="steelblue")
ax2.tick_params(axis='y', labelcolor="steelblue")

# Titre et mise en page
plt.title("Consommation de gaz et tempÃ©rature moyenne par mois")
fig.tight_layout()
plt.show()
```

## Consomation de Gaz par mois

On voit sur le graphique que les **fortes consommations de gaz** se situent durant les mois les plus froids, notamment en **Janvier, FÃ©vrier, Novembre et DÃ©cembre.**

Ã€ lâ€™inverse, la consommation **diminue fortement** pendant les mois les plus chauds, câ€™est-Ã -dire en **Juin, Juillet et AoÃ»t.**

On observe Ã©galement que plus la courbe des **tempÃ©ratures est basse**, plus la **consommation augmente**.
