```python
df['datetime_hour'] = pd.to_datetime(df['datetime_hour'])
df.info()
```

```python
df.plot(x="temperature_2m", y="conso_gaz_mw", kind="scatter")
```

```python
# Corr√©lation entre Temp√©rature et Consommation
correlation = df['temperature_2m'].corr(df['conso_gaz_mw'])
print("Corr√©lation :", correlation)
```

# Corr√©lation entre Temp√©rature et Consomation Gaz :

La corr√©lation est de -0. 53 ce qui signifie que plus le temp√©rature baisse plus la consomation de Gaz augmente 
, probablement en raison de l‚Äôutilisation du chauffage.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[['temperature_2m']].values  # Variable explicative (2D)
y = df['conso_gaz_mw'].values        # Variable √† expliquer

model = LinearRegression()
model.fit(X, y)

print("Coefficient (impact temp√©rature) :", model.coef_[0])
print("Intercept :", model.intercept_)
print("Score R¬≤ :", model.score(X, y))
```

## R√©gression lin√©aire simple :

Co√©fficient : -261 ce qui signifie que pour chaque augmentation de 1¬∞C la consomation baisse en moyenne de 261 Mw


Intercept : La valeur estim√©e de la consomation quand la temp√©rature est de 0¬∞C est de 7606 Kw

Le Score R^2 indique de 27% de la variance de la consomation est expliqu√©e par la temp√©rature

```python
import matplotlib.pyplot as plt
import pandas as pd

# S'assurer que datetime est bien au format datetime
df["datetime"] = pd.to_datetime(df["datetime"])
df["month"] = df["datetime"].dt.month


conso_gaz_moyenne = df.groupby("month")["conso_gaz_mw"].mean()
temp_moyenne = df.groupby("month")["temperature_2m"].mean()


fig, ax1 = plt.subplots(figsize=(10, 5))


ax1.bar(conso_gaz_moyenne.index, conso_gaz_moyenne, color="lightcoral", label="Conso Gaz (MW)")
ax1.set_xlabel("Mois")
ax1.set_ylabel("Consommation de gaz (MW)", color="lightcoral")
ax1.tick_params(axis='y', labelcolor="lightcoral")
ax1.set_xticks(range(1, 13))
ax1.set_xticklabels(["Jan", "F√©v", "Mar", "Avr", "Mai", "Juin", "Juil", "Ao√ªt", "Sept", "Oct", "Nov", "D√©c"])

# Courbe pour la temp√©rature 
ax2 = ax1.twinx()
ax2.plot(temp_moyenne.index, temp_moyenne, color="steelblue", marker="o", label="Temp√©rature (¬∞C)")
ax2.set_ylabel("Temp√©rature moyenne (¬∞C)", color="steelblue")
ax2.tick_params(axis='y', labelcolor="steelblue")

# Titre et mise en page
plt.title("Consommation de gaz et temp√©rature moyenne par mois")
fig.tight_layout()
plt.show()
```

## Consomation de Gaz par mois

On voit sur le graphique que les **fortes consommations de gaz** se situent durant les mois les plus froids, notamment en **Janvier, F√©vrier, Novembre et D√©cembre.**

√Ä l‚Äôinverse, la consommation **diminue fortement** pendant les mois les plus chauds, c‚Äôest-√†-dire en **Juin, Juillet et Ao√ªt.**

On observe √©galement que plus la courbe des **temp√©ratures est basse**, plus la **consommation augmente**.

"Les r√©gions aux climats plus froids ou √† forte densit√© d√©mographique tendent √† consommer davantage de gaz, en particulier lorsque les temp√©ratures chutent."

```python
ensoleillement_moyen = df.groupby("insee_region")["sunshine_duration"].mean()

# Trier les r√©gions de la moins ensoleill√©e √† la plus ensoleill√©e
regions_moins_soleil = ensoleillement_moyen.sort_values()

# Affichage
print("R√©gions les moins ensoleill√©es :")
print(regions_moins_soleil.head(5))  # Les 5 moins ensoleill√©es
```

```python
consomation_elec = df.groupby("insee_region")["conso_elec_mw"].mean()

# Trier les r√©gions de la moins ensoleill√©e √† la plus ensoleill√©e
consomation_elec = ensoleillement_moyen.sort_values()

# Affichage
print("R√©gions avec le plus de consomation √©l√®ctrique :")
print(consomation_elec.head(5)) 
```
